```python
import numpy as np

array = np.random.randint(3, 99, size=(18, 16))
print(array)
```

```python
def find_closest_value(z, A):
    flat_A = A.flatten()
    closest_value = min(flat_A, key=lambda x: abs(x - z)) 
    return closest_value

z = 3

result = find_closest_value(z, array)
print(f"The number is close to {z}: {result}")
```

```python
mean_value = np.mean(array)
median_value = np.median(array)
print(f"Avarege value: {mean_value}")
print(f"Median: {median_value}")
```

```python
import matplotlib.pyplot as plt

plt.imshow(array, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
initial_state = np.copy(array)

for _ in range(10):
    i, j = np.random.randint(18), np.random.randint(16)
    array[i, j] = 0
    
print("Matrix with some null values:")
print(array)
plt.imshow(array, cmap='viridis')
plt.colorbar()
plt.show()

array = np.copy(initial_state)

print("\nRestored array:")
print(array)
plt.imshow(array, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
min_val = np.min(array)
max_val = np.max(array)
normalized_array = (array - min_val) / (max_val - min_val)
print("Normalised array:")
print(normalized_array)
```
